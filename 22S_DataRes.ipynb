```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read CSV Files

df_2017 = pd.read_csv("Statcast_2017.csv")
df_2018 = pd.read_csv("Statcast_2018.csv")
df_2019 = pd.read_csv("Statcast_2019.csv")
df_2020 = pd.read_csv("Statcast_2020.csv")
df_2021 = pd.read_csv("Statcast_2021.csv")
```

```python
#Group 4-seam fastball average velocity by month for each year

vel_2017 = df_2017[df_2017.pitch_type == "FF"].groupby(df_2017["game_date"].str[5:7])["release_speed"].mean().reset_index()
vel_2018 = df_2018[df_2018.pitch_type == "FF"].groupby(df_2018["game_date"].str[5:7])["release_speed"].mean().reset_index()
vel_2019 = df_2019[df_2019.pitch_type == "FF"].groupby(df_2019["game_date"].str[5:7])["release_speed"].mean().reset_index()
vel_2020 = df_2020[df_2020.pitch_type == "FF"].groupby(df_2020["game_date"].str[5:7])["release_speed"].mean().reset_index()
vel_2021 = df_2021[df_2021.pitch_type == "FF"].groupby(df_2021["game_date"].str[5:7])["release_speed"].mean().reset_index()
```

```python
#Change indices so that months match up since seasons have different start and end times

vel_2017.index = [1, 2, 3, 4, 5, 6, 7]
vel_2020.index = [4, 5, 6]
vel_2021.index = [1, 2, 3, 4, 5, 6, 7]
```

```python
#Plot average velocity by month for each MLB season 2017-2021

fig, ax = plt.subplots()

vel_2017.plot(x = "game_date", ax = ax, style = "o-", label = 2017)
vel_2018.plot(ax = ax, style = "o-")
vel_2019.plot(ax = ax, style = "o-")
vel_2020.plot(ax = ax, style = "o-")
vel_2021.plot(ax = ax, style = "o-")
plt.grid(axis = "y")
plt.legend([2017, 2018, 2019, 2020, 2021])
plt.xlabel("Month")
ax.set_xticklabels(["", "Mar.", "Apr.", "May", "Jun.", "Jul.", "Aug.", "Sept.", "Oct."])
plt.ylabel("Velocity in MPH")
plt.title("Average Four-Seam Fastball Velocity by Month")

#Delete symbol to download graph
#plt.savefig('22S_DataRes_velocity.png')
```

```python
#Group 4-seam fastball average spin rate by month for each year

spin_2017 = df_2017[df_2017.pitch_type == "FF"].groupby(df_2017["game_date"].str[5:7])["release_spin_rate"].mean().reset_index()
spin_2018 = df_2018[df_2018.pitch_type == "FF"].groupby(df_2018["game_date"].str[5:7])["release_spin_rate"].mean().reset_index()
spin_2019 = df_2019[df_2019.pitch_type == "FF"].groupby(df_2019["game_date"].str[5:7])["release_spin_rate"].mean().reset_index()
spin_2020 = df_2020[df_2020.pitch_type == "FF"].groupby(df_2020["game_date"].str[5:7])["release_spin_rate"].mean().reset_index()
spin_2021 = df_2021[df_2021.pitch_type == "FF"].groupby(df_2021["game_date"].str[5:7])["release_spin_rate"].mean().reset_index()
```

```python
#Change indices so that months match up since seasons have different start and end times

spin_2017.index = [1, 2, 3, 4, 5, 6, 7]
spin_2020.index = [4, 5, 6]
spin_2021.index = [1, 2, 3, 4, 5, 6, 7]
```

```python
#Plot average spin rate by month for each MLB season 2017-2021

fig, ax2 = plt.subplots()

spin_2017.plot(ax = ax2, style = "o-")
spin_2018.plot(ax = ax2, style = "o-")
spin_2019.plot(ax = ax2, style = "o-")
spin_2020.plot(ax = ax2, style = "o-")
spin_2021.plot(ax = ax2, style = "o-")
plt.grid(axis = "y")
plt.legend([2017, 2018, 2019, 2020, 2021])
plt.xlabel("Month")
ax2.set_xticklabels(["", "Mar.", "Apr.", "May", "Jun.", "Jul.", "Aug.", "Sept.", "Oct."])
plt.ylabel("Spin Rate in RPM")
plt.title("Average Four-Seam Fastball Spin Rate by Month")

#Delete symbol to download graph
#plt.savefig('22S_DataRes_spin')
```
